```python
import pandas as pd
```

```python
data_2022 = pd.read_csv('2022_results.csv')
data_2023 = pd.read_csv('2023_results.csv')
```

```python
games_played_2023 = 20
total_games_in_season = 162  # Modify this if the total number of games in the season is different
progress = games_played_2023 / total_games_in_season

weight_2023 = progress
weight_2022 = 1 - weight_2023

data_2022 = data_2022.drop(['home_is_winner', 'game_date', 'home_tm', 'away_tm'], axis=1) * weight_2022
data_2023 = data_2023.drop(['home_is_winner', 'game_date', 'home_tm', 'away_tm'], axis=1) * weight_2023
#weighted_data = data_2022.add(data_2023)
```
