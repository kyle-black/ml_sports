```python
def calculate_bets(df):
    # Initialize bankroll
    bankroll = 100

    # New columns for bet_amount, bet_net, and updated bankroll
    df['bet_amount'] = 0
    df['bet_net'] = 0
    df['bankroll'] = 0

    for idx, row in df.iterrows():
        if row['bethome'] or row['betaway']:  # Only place a bet if one of these is True
            bet = bankroll * 0.05  # 5% of current bankroll
            net = 0
            if row['bethome'] and row['home_is_winner'] == 1:
                net = bet + bet * (100 / abs(row['home_odds']) if row['home_odds'] < 0 else row['home_odds'] / 100)
                bankroll += net
            elif row['betaway'] and row['home_is_winner'] == 0:
                net = bet + bet * (100 / abs(row['away_odds']) if row['away_odds'] < 0 else row['away_odds'] / 100)
                bankroll += net
            else:  # If neither condition above is met, the bet is lost
                net = -bet
                bankroll -= bet

            # Update the DataFrame
            df.at[idx, 'bet_amount'] = bet
            df.at[idx, 'bet_net'] = net
        # If neither bethome nor betaway is true, no bet is placed and bet_amount, bet_net remain 0
        df.at[idx, 'bankroll'] = bankroll

    return df

# Apply the function to your DataFrame
df = calculate_bets(df)
```
